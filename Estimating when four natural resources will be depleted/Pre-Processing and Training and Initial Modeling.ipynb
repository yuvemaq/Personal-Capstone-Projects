**Data Description**

Along with description from other sites such as [PDFCOFEE](https://pdfcoffee.com/bwla-porosity-logs-pdf-free.html) and [AAP WIKI](https://wiki.aapg.org/Well_log_analysis_for_reservoir_characterization), we'll describe each parameter in depth.

- **Depth**: Depth of well with every 0.5 m reading
- **ILD**: Measures the resistivity of the fluid in formation in the Virgin Zone (or the zone not commingled with the drilling Zone) 
- RLL3: Laterlog 3 data
- **SP**: spontaneous data; **Uses**: Lithology interpretation, Rw and Rwe calculation, detect permeable zone, etc.
- **TT1**: Measures interval transit time as it goes through the formation 
- **CALD**: Measures size of the hole
- **RHOB**: Bulk density data (Units: g/cm3, kg/m3)
- **RHOC**: Corrected bulk density
- **NPLS**: Compensated Neutron Porosity
- **DPLS**: Density Porosity 
- **GR**: Similar to spontaneous potential; **Uses:** Determine where sandstone or shale is

Spontaneous potential is the result of an electrical potential which is created by co-mingling of drilling fluid with formation fluid. Gammay ray measures natural gamma radiation of the Earth; shale tends to hold radioactive material (whereas in sandstone the radioactivity is washed out).

## Monte Carlo Simulation

The essence of a Monte Carlo simulation is it is simulations evolving randomly. In the law of large numbers, an average tends to come closer to its expected value the more samples there are. In Monte Carlo simulation, we want to get an unbiased representative group of samples from a large pool if we allow simulation to evolve randomly. One relevant use for this stochastic tool is for hydrocarbon resource estimation (probabilistic hydrocarbon reserves estimation). In this probabilistic method, a fit distribution curve for each parameter is introduced and through a Monte Carlo simulation, a distribution curve for the result (reserves) can be developed. 

An equation for calculating oil reserve is as follows: 

**N = [A*h**(1-Sw)/B0]**

- N = volumetric oil reserve (or OOIP)
- A = area (m2)
- h = Thickness (or net pay thickness, m)
-  = porosity (fraction)
- Sw = water saturation
- B0 = Formation Volume Factor (RB/STB) 

In Monte Carlo simulator these most likely, minimum possible and maximum possible values for each parameter is entered. After that when simulator is initiated it generates 100/500/1000 values (depending up on the iteration level) based on random number generator algorithm for each parameter within the range specified and creates a probabilistic picture of the volumetric reserve estimation. Finally simulator provides P90, P50, P25, P10 values of reserve or what ever probabilistic value of reserve we want.

Some unit definitions before we get started:

- BBO: billion barrels of oil
- scf/bbl: standard cubic foot per barrel

MDD is the maximum loss from a peak to a trough before a new peak is produced. One use of this is to measure a downside risk over a period of time. 

MDD = (Trough Val - Peak Val)/Peak Val

**Past Performance to Predict Future**

Here, we will try to estimate the expected level of production (mu), using historic production (barrels_daily) assuming past mean levelsn will continue into the future. 

```python
#calculate the compound annual growth rate (CAGR) which 
#will give us our mean return input (mu) 
days = (venez_ds.index[-1] - venez_ds.index[0]).days
cagr = ((((venez_ds['1'][-1]) / venez_ds['1'][1])) ** (365.0/days)) - 1
mu = cagr
print ('CAGR =',str(round(cagr,4)*100)+"%")
```

```python
from math import sqrt, fabs

# Make a series of percentage production and calculate 
# the annual volatility of production.
# Higher volatility = less reliable production data
venez_ds['Production'] = venez_ds['1'].pct_change()
vol = venez_ds['Production'].std()*sqrt(252)
print ("Annual Volatility =",str(round(vol,4)*100)+"%")
```

```python
#set up empty list to hold our ending values for each simulated production series
result = []
 
#Define Variables
start_prod = venez_ds['1'][-1] #starting production (i.e. last available real oil production)
trading_days = 252 #Number of trading days

 
#choose number of runs to simulate - I have chosen 1,00
for i in range(100):
    
    #create list of daily returns using random normal distribution
    daily_prod=np.random.normal(mu/trading_days,vol/sqrt(trading_days),trading_days)+1
 
    #set starting production and create production series generated by above random daily production
    prod_list = [start_prod]
 
    for x in daily_prod:
        prod_list.append(prod_list[-1]*x)
 
    #plot data from each individual run which we will plot at the end
    plt.plot(prod_list)
 
    #append the ending value of each simulated run to the empty list we created at the beginning
    result.append(prod_list[-1])

plt.rcParams["figure.figsize"] = (18,12) 
plt.suptitle('Monte Carlo Simulation', fontsize=20)
#show the plot of multiple price series created above
plt.show()
 
#create histogram of ending stock values for our mutliple simulations
plt.suptitle('Ending Production Values', fontsize=20)
plt.hist(result,bins=50)

plt.show()
```

**Simulated Results with Long and Short moving average**

In many methods that perform well in backtests, it can perform poorly in future methods. We can run multiple simulations and look fruture for trends that emerge. This is similar to the Monte Carlo simulation as above, except we will include code for Spot instances. 

```python
# Initialize the short and long windows
short_window = 5
long_window = 30

# Number of simulations 
sim_num = 3

# Set the initial barrels_daily
initial_prod = float(100000.0)

# Create a DataFrame `positions`
positions = pd.DataFrame(index=venez_ds.index).fillna(0.0)

#set up empty list to hold our ending values for each simulated production series
sim_result = []

# Set up empty list to hold protfolio value for each simulated production serries, this is the value of position['total']
portfolio_total = []
 
#Define Variables
start_prod = venez_ds['1'][-1] #starting stock price (i.e. last available real stock price)
trading_days = 252 # Number of trading days



#calculate the compound annual growth rate (CAGR) which 
#will give us our mean return input (mu) 
days = (venez_ds.index[-1] - venez_ds.index[0]).days
cagr = ((((venez_ds['1'][-1]) / venez_ds['1'][1])) ** (365.0/days)) - 1
mu = cagr
 
#create a series of percentage returns and calculate 
#the annual volatility of returns. Generally, the higher the volatility, 
#the riskier the investment in that stock, which results in investing in one over another.
venez_ds['Production'] = venez_ds['1'].pct_change()
vol = venez_ds['Production'].std()*sqrt(trading_days)


# Initialize the `signals` DataFrame 
signals = pd.DataFrame(index=venez_ds.index)

# initialize by setting the value for all rows in this column to 0.0.
signals['signal'] = 0.0
signals['short_mavg'] = 0.0


#choose number of runs to simulate - I have chosen 1,000
for i in range(sim_num):
    #create list of daily returns using random normal distribution
    daily_prod = np.random.normal(mu/trading_days,vol/sqrt(trading_days),trading_days)+1
 
    #set starting price and create price series generated by above random daily returns
    prod_list = [start_prod]
 
    for x in daily_prod:
        prod_list.append(prod_list[-1]*x)
 
    #plot data from each individual run which we will plot at the end
    plt.plot(prod_list)
    
    #Convert list to Pandas DataFrame
    df_prod_list = pd.DataFrame(prod_list)
   
    #append the ending value of each simulated run to the empty list we created at the beginning
    sim_result.append(prod_list[-1])
    
    # Create short simple moving average over the short & long window
    signals['short_mavg']  = df_prod_list[0].rolling(short_window).mean()
    signals['long_mavg']  = df_price_list[0].rolling(long_window).mean()
 
    # Create a signal when the short moving average crosses the long moving average, 
    # but only for the period greater than the shortest moving average window. 
    signals['signal'][short_window:] = np.where(signals['short_mavg'][short_window:] 
                                            > signals['long_mavg'][short_window:], 1.0, 0.0)   

    # Generate trading orders
    signals['positions'] = signals['signal'].diff()
    
    #plot data from each individual run which we will plot at the end
    plt.plot(prod_list)
    plt.plot(signals['short_mavg'])
    plt.plot(signals['long_mavg'])
    
    # Buy 100 shares
    positions[venez_ds] = 100*signals['signal']
  
    # Initialize the portfolio with value owned   
    portfolio = positions.multiply(df_prod_list[0], axis=0)

    # Store the difference in shares owned 
    pos_diff = positions.diff()

    # Add `holdings` to portfolio
    portfolio['extracted'] = (positions.multiply(df_prod_list[0], axis=0)).sum(axis=1)

    # Add `cash` to portfolio
    portfolio['reserve estimation'] = initial_prod - (pos_diff.multiply(df_prod_list[0], axis=0)).sum(axis=1).cumsum()   

    # Add `total` to portfolio
    portfolio['total'] = portfolio['reserve estimation'] + portfolio['extracted']
    
    #append the ending value of each simulated run to the empty list we created at the beginning
    portfolio_total.append(portfolio['total'].iloc[-1])

         
# Show the plot
plt.rcParams["figure.figsize"] = (15,18)
plt.suptitle('Simulated Results with long and short moving avg', fontsize=20)
plt.show()
    
#create histogram of ending stock values for our mutliple simulations
plt.rcParams["figure.figsize"] = (9,8)    
plt.suptitle('Monte Carlo Simulation Production', fontsize=20)
plt.hist(sim_result,bins=50)
plt.show()

#create histogram of ending protfolio values for our mutliple simulations, compare this value to "initial_capital"
# Clean portfolio_total from NaN
portfolio_total = [x for x in portfolio_total if x != 'nan']
plt.hist(portfolio_total,bins=50)
plt.suptitle('\Simulated Results (based on $100K investment)', fontsize=20)
plt.show()
```
