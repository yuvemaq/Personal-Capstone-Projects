### Well 110

```python
# plotting with de-spiking
# Plot Gamma ray and DPHI-NPHI well logs
segm_GR = well110.data['GR'].smooth(window_length=10, samples=False)
segm_NPHI = well110.data['NPHI'].smooth(window_length=10, samples=False)
segm_DPHI = well110.data['DPHI'].smooth(window_length=10, samples=False)
segm_ILD = well110.data['ILD'].smooth(window_length=10, samples=False)

# Matplotlib plotly used with welly objects
fig, axes = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(5,10))
fig.subplots_adjust(wspace=0.1)
#----------------------------------------------------------------------------1st track------------------------
axes[0].clear()
segm_GR.plot(ax= axes[0], c='g')
axes[0].set_xlim([0,150])
axes[0].fill_betweenx(segm_GR.basis, 150, segm_GR, hatch='///',edgecolor='green', facecolor='none')
axes[0].grid()
#---------------------------------------------------------------------------2nd track-------------------------
ax1 = axes[1]
ax2 = ax1.twiny()
ax1.set_xlim([0.18, 0.7])
ax2.set_xlim([0.0, 0.7])
segm_NPHI.plot(ax=ax1, c='r')
segm_DPHI.plot(ax=ax2, c='b')
segm_2 = -segm_NPHI + 0.50
segm_2.plot(ax=ax2, c='None')
ax2.fill_betweenx(segm_NPHI.basis, segm_DPHI, segm_2, color='red', alpha=0.5, where=(segm_2>segm_DPHI))
axes[1].grid()
#---------------------------------------------------------------------------3rd track-------------------------
segm_ILD.plot(ax= axes[2], c='y')
axes[2].grid()
```
