### Rolling OLS

What I want to do next is Rolling ordinary least square, which applies OLS across a fixed window of observations and then moves or slides that window across the data set. The 'window' determines the number of observations used in each OLS regression.

```python
# Change column names to target, feature0 and feature1
ds_ppm_analysis2.rename(columns = {"Acute respiratory symptoms avoided due to particulate matter [PM2.5] removed (cases/yr)":"target"}, inplace=True)
ds_ppm_analysis2.rename(columns = {"Particulate matter [PM2.5] (kg/yr)":"feature0"}, inplace=True)
ds_ppm_analysis2.rename(columns = {"Average reduction in nighttime ambient temperature (Celsius)":"feature1"}, inplace=True)
```

```python
# Fit model using formula and a window of 25. -1 suppresses the intercept
roll_reg = RollingOLS.from_formula('target ~ feature0 + feature1 -1', window=25, data=ds_ppm_analysis2)
model = roll_reg.fit()
```

```python
# Check parameters
model.params[20:30] # No params between 0-23 due to window=25
```

```python
fig = model.plot_recursive_coefficient(variables=['feature0'])
plt.xlabel('Time step')
plt.ylabel('Coefficient value')
plt.figure(figsize=(40,40))
plt.show()
```

Next, I'll try recursive least squares, or expanding window rolling regression. It is *expanding* rather than *moving*.

### Recursive LS

```python
reg_rls = sm.RecursiveLS.from_formula(
    'target ~ feature0 + feature1 -1', ds_ppm_analysis2)
model_rls = reg_rls.fit()
print(model_rls.summary())
```

```python
# Check how the values of coefficients changed in real time
fig = model_rls.plot_recursive_coefficient(range(reg_rls.k_exog), legend_loc='upper right')
ax_list = fig.axes
for ax in ax_list:
    ax.set_xlim(0, None)
ax.figure.set_dpi(100)
ax_list[-1].set_xlabel('Time step')
ax_list[0].set_title('Coefficient value');
```

### Rolling OLS for Tampa

```python
# Change column names to target, feature0 and feature1
ds_ppm_analysis4.rename(columns = {"Acute respiratory symptoms avoided due to particulate matter [PM2.5] removed (cases/yr)":"target"}, inplace=True)
ds_ppm_analysis4.rename(columns = {"Particulate matter [PM2.5] (kg/yr)":"feature0"}, inplace=True)
ds_ppm_analysis4.rename(columns = {"Average reduction in nighttime ambient temperature (Celsius)":"feature1"}, inplace=True)
```

```python
# Fit model using formula and a window of 25. -1 suppresses the intercept
roll_reg3 = RollingOLS.from_formula('target ~ feature0 + feature1 -1', window=25, data=ds_ppm_analysis4)
model3 = roll_reg3.fit()
```

```python
fig3 = model3.plot_recursive_coefficient(variables=['feature0'])
plt.xlabel('Time step')
plt.ylabel('Coefficient value')
plt.figure(figsize=(40,40))
plt.show()
```

### Recursive LS for Tampa

```python
reg_rls3 = sm.RecursiveLS.from_formula(
    'target ~ feature0 + feature1 -1', ds_ppm_analysis4)
model_rls3 = reg_rls3.fit()
print(model_rls3.summary())
```

```python
# Check how the values of coefficients changed in real time
fig3 = model_rls3.plot_recursive_coefficient(range(reg_rls3.k_exog), legend_loc='upper right')
ax_list3 = fig3.axes
for ax3 in ax_list:
    ax3.set_xlim(0, None)
ax3.figure.set_dpi(100)
ax_list3[-1].set_xlabel('Time step')
ax_list3[0].set_title('Coefficient value');
```

### Multiple Linear Regression for Tampa Data - Model 1

Since we've seen that multiple linear regression was the best model, we will use only that for the Tampa data.

```python
x5 = ds_ppm_analysis4.drop(["target"], axis=1)

# Create constants for X, so the model knows its bounds
x5 = sm.add_constant(x5)

# Split the data
y_tpa = ds_ppm_analysis4[["target"]]
x5_train, x5_test, y_train_tpa, y_test_tpa = train_test_split(x5, y_tpa, test_size=0.25, shuffle=True)
```

```python
rModel_5 = sm.OLS(y_train_tpa, x5_train)

# Fit the model
fit_model5 = rModel_5.fit()

# Evaluate summary
fit_model5.summary()
```

```python
# Use our Model 1 to make predictions
y_predict_tpa = fit_model5.predict(x5_test)
```

```python
# Plot the predictions
_ = plt.scatter(y_test_tpa, y_predict_tpa)
_ = plt.title("Model 1 vs. Actual")
_ = plt.xlabel("Actual No. of ARS avoided due to PM2.5 removed (cases/yr)")
_ = plt.ylabel("Predicted No. ARS avoided due to PM2.5 removed (cases/yr)")
# Add a line for perfect correlation
_ = plt.plot([x for x in range(0,4)], [x for x in range(0,4)], color = "orange")
plt.show()
```

#### Plotting residuals and heteroscedasticity for Tampa data

```python
residual_tpa = y_train_tpa - y_predict_tpa

#define figure size
fig2 = plt.figure(figsize=(20,10))

#produce regression plots
fig_tpa = sm.graphics.plot_regress_exog(fit_model5, "feature0", fig=fig2)
```
